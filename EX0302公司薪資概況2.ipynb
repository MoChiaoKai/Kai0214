```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}

df = pd.DataFrame(data)
df
```

```python
df['Salary_USD'] = df['Salary'] / 30
df
```

```python
def categorize_salary(salary):
    if salary < 60000:
        return 'Low'
    elif 60000 <= salary < 75000:
        return 'Medium'
    else:
        return 'High'

df['Salary_Category'] = df['Salary'].apply(categorize_salary)
df
```

```python
pivot_count = pd.pivot_table(df, values='ID', index='Department', columns='City', aggfunc='count')
print(pivot_count)
```

```python
pivot_mean_salary = pd.pivot_table(df, values='Salary', index='Department', columns='City', aggfunc='mean')
print(pivot_mean_salary)
```
